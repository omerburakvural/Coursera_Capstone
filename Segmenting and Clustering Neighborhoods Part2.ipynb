```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import geocoder
```

<h3>Scrape the wikipedia page</h3>

```python
wiki_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
wiki_df.columns = ["PostalCode", "Borough", "Neighbourhood"]
```

<h3>Clean the Dataframe</h3>

```python
wiki_df = wiki_df[wiki_df['Borough'] != 'Not assigned']
```

```python
#remove duplicates
wiki_df = wiki_df.drop_duplicates()
#update the index
if(wiki_df.index.name != 'PostalCode'):
  wiki_df = wiki_df.set_index('PostalCode')
```

```python
#Replacin meaningless data
wiki_df['Neighbourhood'].replace("Not assigned", wiki_df["Borough"],inplace=True)
```

```python
#importing location data
geo_df = pd.read_csv("https://raw.githubusercontent.com/omerburakvural/Coursera_Capstone/main/Geospatial_Coordinates.csv")
geo_df.columns = ["PostalCode", "Latitude", "Longitude"]
geo_df.head()
```

```python
if(geo_df.index.name != 'PostalCode'):
    geo_df = geo_df.set_index('PostalCode')
    
geo_df.head()
```

```python
wiki_df = wiki_df.join(geo_df)
wiki_df.head(10)
```
